```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
data = {
    'Día': ['viernes 6 de octubre', 'sábado 7 de octubre', 'domingo 8 de octubre', 'lunes 9 de octubre',
            'martes 10 de octubre', 'miércoles 11 de octubre', 'jueves 12 de octubre', 'Promedio'],
    'Tiempo en Redes Sociales': [249, 157, 260, 189, 335, 386, 373, 278],
    'Tiempo en Apps audiovisuales': [69, 43, 73, 103, 86, 99, 95, 81],
    'Tiempo en Apps de Ocio': [64, 45, 45, 20, 33, 92, 62, 52],
    'Tiempo Total en pantalla': [452, 365, 473, 345, 511, 685, 636, 495]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Crear el gráfico de barras para el tiempo total en pantalla
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Día'], df['Tiempo Total en pantalla'], color='lightblue')
plt.xlabel('Día')
plt.ylabel('Tiempo (minutos)')
plt.title('Tiempo Total en Pantalla')
plt.xticks(rotation=45)

# Agregar etiquetas a las barras
for bar, minutes in zip(bars, df['Tiempo Total en pantalla']):
    plt.text(bar.get_x() + bar.get_width() / 2, minutes, str(minutes), ha='center', va='bottom')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
data = {
    'Día': ['viernes 6 de octubre', 'sábado 7 de octubre', 'domingo 8 de octubre', 'lunes 9 de octubre',
            'martes 10 de octubre', 'miércoles 11 de octubre', 'jueves 12 de octubre', 'Promedio'],
    'Tiempo en Redes Sociales': [249, 157, 260, 189, 335, 386, 373, 278],
    'Tiempo en Apps audiovisuales': [69, 43, 73, 103, 86, 99, 95, 81],
    'Tiempo en Apps de Ocio': [64, 45, 45, 20, 33, 92, 62, 52],
    'Tiempo Total en pantalla': [452, 365, 473, 345, 511, 685, 636, 495]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Crear visualizaciones
fig, ax = plt.subplots(3, 1, figsize=(10, 18))

# Función para agregar etiquetas en las barras
def add_bar_labels(ax):
    for p in ax.patches:
        ax.annotate(f'{p.get_height()} min', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Gráfico de barras para el tiempo en redes sociales
df.plot(x='Día', y='Tiempo en Redes Sociales', kind='bar', color='skyblue', ax=ax[0])
ax[0].set_xlabel('Día')
ax[0].set_ylabel('Tiempo (minutos)')
ax[0].set_title('Tiempo en Redes Sociales')
add_bar_labels(ax[0])

# Gráfico de barras para el tiempo en aplicaciones audiovisuales
df.plot(x='Día', y='Tiempo en Apps audiovisuales', kind='bar', color='lightcoral', ax=ax[1])
ax[1].set_xlabel('Día')
ax[1].set_ylabel('Tiempo (minutos)')
ax[1].set_title('Tiempo en Apps Audiovisuales')
add_bar_labels(ax[1])

# Gráfico de barras para el tiempo en aplicaciones de ocio
df.plot(x='Día', y='Tiempo en Apps de Ocio', kind='bar', color='lightgreen', ax=ax[2])
ax[2].set_xlabel('Día')
ax[2].set_ylabel('Tiempo (minutos)')
ax[2].set_title('Tiempo en Apps de Ocio')
add_bar_labels(ax[2])

# Ajustes de diseño
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
data = {
    'Día': ['viernes 6 de octubre', 'sábado 7 de octubre', 'domingo 8 de octubre', 'lunes 9 de octubre',
            'martes 10 de octubre', 'miércoles 11 de octubre', 'jueves 12 de octubre'],
    'Tiempo en Redes Sociales': [249, 157, 260, 189, 335, 386, 373],
    'Tiempo en Apps audiovisuales': [69, 43, 73, 103, 86, 99, 95],
    'Tiempo en Apps de Ocio': [64, 45, 45, 20, 33, 92, 62],
    'Tiempo Total en pantalla': [452, 365, 473, 345, 511, 685, 636]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# Configurar el gráfico
plt.figure(figsize=(10, 6))

# Crear un gráfico de líneas para cada categoría
plt.plot(df['Día'], df['Tiempo en Redes Sociales'], marker='o', label='Redes Sociales', linestyle='-', color='skyblue')
plt.plot(df['Día'], df['Tiempo en Apps audiovisuales'], marker='o', label='Apps Audiovisuales', linestyle='-', color='lightcoral')
plt.plot(df['Día'], df['Tiempo en Apps de Ocio'], marker='o', label='Apps de Ocio', linestyle='-', color='lightgreen')

# Configurar el eje y
plt.ylabel('Tiempo (minutos)')

# Título y etiquetas
plt.title('Uso de Tiempo a lo Largo de los Días')
plt.xlabel('Día')
plt.xticks(rotation=45)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
